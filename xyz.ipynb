```python
def convet_to_gender(f):
    if ("mostly" in f):
        words = f.split("mostly_")
        return words[1]
    else:
        return f
```

```python
for index, row in df2.iterrows():
    if (row['gender'] is NaN):
        artist = row['artist']

        if ("featuring" in artist):
            words = artist.split()
            artist_ = words[0]

            guess = convet_to_gender(d.get_gender(artist_))
            if(guess == "male"):
                df2.at[index,"gender"] = "male"
            elif (guess == "female"):
                df2.at[index,"gender"] = "female"

        elif ("&" in artist):
            artists = artist.split(" & ")
            try:
                artist_1 = artists[0]
                artist_2 = artists[1]

                guess_1 = convet_to_gender(d.get_gender(artist_1))
                guess_2 = convet_to_gender(d.get_gender(artist_2))

                if (guess_1 == "male" and guess_2 == "male"):
                    df2.at[index,"gender"] = "male"
                elif (guess_1 == "female" and guess_2 == "female"):
                    df2.at[index,"gender"] = "female"
                elif ((guess_1 == "male" and guess_2 == "unknown") or (guess_1 == "unknown" and guess_2 == "male")):
                    df2.at[index,"gender"] = "male"
                elif ((guess_1 == "female" and guess_2 == "unknown") or (guess_1 == "unknown" and guess_2 == "female")):
                    df2.at[index,"gender"] = "female"
            except:
                print(artists)

    if (row['gender'] is not NaN):
        print(row['gender'])
```
