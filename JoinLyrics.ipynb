```python
import pandas as pd
```

```python
df = pd.read_csv('lyrics_public.csv')
```

```python
df['Song'] = df['Song'].str.lower()
df['SongID'] = df['SongID'].str.lower()
df.head()
df
```

```python
hot = pd.read_csv('hot100_cleaned.csv')
```

```python
hot.head()
hot
```

```python
hot['title'] = hot['title'].str.lower()
hot['artist'] = hot['artist'].str.lower()
hot['key'] = hot['title'] + hot['artist']
hot.head()
```

```python
merged = pd.merge(hot,df,left_on='key',right_on='SongID',how='left')
merged.to_csv('song_lyrics.csv')
```
