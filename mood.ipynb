```python
import pandas as pd
```

```python
df = pd.read_csv('topics_4_1000_final_3Nov.csv')
```

```python
df.head()
df.columns
```

```python
extract = df[['title_x', 'artist_x', 'year_x', 'key', 'MTLD', 'TTR', 'CTTR',
       'sentimentScore', 'sentimentScore_pos', 'sentimentScore_neg',
       'emo_words', 'emoWordCount', 'joy_normalized',
       'sadness_normalized','anger_normalized',
       'disgust_normalized','trust_normalized',
       'anticipation_normalized','fear_normalized',
       'surprise_normalized', 'emo_score', 'happy_normalized',
       'sorrow_normalized','topic']]
extract['decade'] = extract['year_x'].apply(lambda x: str(x)[2]+'0s')
```

```python
extract
```

```python
extract.isna().sum()
```

```python
extract.dropna(inplace=True)
```

```python
extract.isna().sum()
```

```python
extract.head()
extract.columns
```

```python
mtld = extract.groupby(by='decade')['MTLD'].describe()
mtld = mtld.reindex(['50s','60s','70s','80s','90s','00s','10s','20s'])
mtld
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")
 


# Plot the responses for different events and regions
sns.lineplot(x="decade", y="mean",
             data=mtld, label="Mean")
sns.lineplot(x="decade", y="75%",
             data=mtld, label="75 Percentile Value")
sns.lineplot(x="decade", y="25%",
             data=mtld, label="25 Percentile Value")
# labels for x & y axis
plt.xlabel('Decades')
plt.ylabel('MTLD Values')
 
# title of plot
plt.title('Lexical Diversity Values over the years')

plt.savefig('mtld-decades.png', dpi=300)
```

```python
anger = extract.groupby(by='decade')['anger_normalized'].describe()
anger = anger.reindex(['50s','60s','70s','80s','90s','00s','10s','20s'])
anger
```

```python
# Plot the responses for different events and regions
sns.lineplot(x="decade", y="mean",
             data=anger, label="Mean")
sns.lineplot(x="decade", y="75%",
             data=anger, label="75 Percentile Value")
sns.lineplot(x="decade", y="25%",
             data=anger, label="25 Percentile Value")
# labels for x & y axis
plt.xlabel('Decades')
plt.ylabel('Anger (Normalized) Values')
 
# title of plot
plt.title('Anger (Normalized) Values over the years')

plt.savefig('anger-decades.png', dpi=300)
```

```python
dff = extract.groupby(by='decade')['disgust_normalized'].describe()
dff = dff.reindex(['50s','60s','70s','80s','90s','00s','10s','20s'])
dff

```

```python
params = ['MTLD', 'TTR', 'CTTR', 'sentimentScore', 'sentimentScore_pos',
       'sentimentScore_neg', 'emoWordCount', 'joy_normalized',
       'sadness_normalized', 'anger_normalized', 'disgust_normalized',
       'trust_normalized', 'anticipation_normalized', 'fear_normalized',
       'surprise_normalized', 'emo_score', 'happy_normalized',
       'sorrow_normalized']
for param in params:
    dff = extract.groupby(by='decade')[param].describe()
    dff = dff.reindex(['50s','60s','70s','80s','90s','00s','10s','20s'])
    plt.clf()
    # Plot the responses for different events and regions
    sns.lineplot(x="decade", y="mean",
                data=dff, label="Mean")
    sns.lineplot(x="decade", y="75%",
                data=dff, label="75 Percentile Value")
    sns.lineplot(x="decade", y="25%",
                data=dff, label="25 Percentile Value")
    # labels for x & y axis
    plt.xlabel('Decades')
    plt.ylabel(param+' Values')
    
    # title of plot
    plt.title(param+' Values over the years')
    # plt.show()
    plt.savefig('spotify-'+param+'-decades.png', dpi=300)
```

```python
params = ['MTLD', 'TTR', 'CTTR', 'sentimentScore', 'sentimentScore_pos',
       'sentimentScore_neg', 'emoWordCount', 'joy_normalized',
       'sadness_normalized', 'anger_normalized', 'disgust_normalized',
       'trust_normalized', 'anticipation_normalized', 'fear_normalized',
       'surprise_normalized', 'emo_score', 'happy_normalized',
       'sorrow_normalized']
topics = [1,2,3,4]
for param in params:
    dff = extract.groupby(by='topic')[param].describe()
    # dff = dff.reindex(['50s','60s','70s','80s','90s','00s','10s','20s'])
    plt.clf()
    # Plot the responses for different events and regions
    sns.lineplot(x=topics, y="mean",
                data=dff, label="Mean")
    sns.lineplot(x=topics, y="75%",
                data=dff, label="75 Percentile Value")
    sns.lineplot(x=topics, y="25%",
                data=dff, label="25 Percentile Value")
    # labels for x & y axis
    plt.xlabel('Topics')
    plt.ylabel(param+' Values')
    plt.xticks(topics)
    # title of plot
    plt.title(param+' Values v/s Topics')
    # plt.show()
    plt.savefig('spotify-'+param+'-topics.png', dpi=300)
```
